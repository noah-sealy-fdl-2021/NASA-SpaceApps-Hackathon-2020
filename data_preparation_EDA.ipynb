```python
dirt = "/Users/zarreennaowalreza/Desktop/SpaceApps/SCISAT/"
csv_files = glob.glob(dirt+'/*.csv') 

file = '/Users/zarreennaowalreza/Desktop/SpaceApps/ACEFTS_L2_v4p1_CO2_636.csv'

std_col = 'std ' + file.split('/')[5].split('ACEFTS_L2_v4p1_')[1].split('.')[0]

cols = ['Alt_Mean', std_col, 'date', 'long', 'lat']

df = pd.read_csv(file, usecols=cols)
df.rename(columns={'Alt_Mean': 'Alt_Mean_CO2_636'}, inplace=True)

for f in csv_files:
    print(f)
    std_col = 'std ' + f.split('/')[6].split('ACEFTS_L2_v4p1_')[1].split('.')[0]

    cols = ['Alt_Mean', std_col, 'date', 'long', 'lat']
    df_tmp = pd.read_csv(f, usecols=cols)
    df_tmp.rename(columns={'Alt_Mean': 'Alt_Mean_'+std_col}, inplace=True)
    df = pd.merge(df, df_tmp,  how='left', on=['date','lat', 'long'])
    print(df.shape)

```
